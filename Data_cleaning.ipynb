```python
import pandas as pd

df = pd.read_excel("c.xlsx")
df
```

```python
# drop duplicates
df = df.drop_duplicates()
```

```python
# Remove unnecessary data
df = df.drop(columns="Not_Useful_Column")
```

```python
# Clean the inputs
# df["Last_Name"] = df["Last_Name"].str.strip('...')
# df["Last_Name"] = df["Last_Name"].str.strip('/')
# df["Last_Name"] = df["Last_Name"].str.strip('_')
```

```python
df["Last_Name"] = df["Last_Name"].str.strip('123._/')
df["Last_Name"]
```

```python
# Clean data with very diff formating (phone nums in this case)
# strip them from characters and then reformat them

# 2) Remove all non-digits
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]', '' , regex=True)
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('<NA->-', '')
df['Phone_Number'] = df['Phone_Number'].replace('--', '')
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--', '')
```

```python
# spliting a column and creating new columns for that
df[['Street Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)

```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df = df.replace('N/a', '')
```

```python
df = df.fillna('') # fills missing values (NaN or None) in your DataFrame or Series with something you specify.
```

```python
# Filtering
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace=True) # modifies df directly, if False, creates a copy

df
```

```python
for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace=True) # modifies df directly, if False, creates a copy

# or the same
# dr.dropna(subset="Phone_Number", inplace=True)
```

```python
# Reset indexing
df.reset_index(drop=True) #drops original index
df
```
